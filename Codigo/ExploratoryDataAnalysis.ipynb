# 3. Búsqueda de datos duplicados y outliers

```python
print("Quantity of duplicated rows")
print(df.duplicated().sum())
```

This time, we don't have to work with duplicated data <br> Now, search for outliers

```python
fig, axes = plt.subplots(6,3)
sns.set(rc={'figure.figsize':(31.7,28.27)})

df_mod = df.copy()
df_mod.drop(labels=["Response","Z_Revenue","Z_CostContact","Complain","Dt_Customer","ID","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5","AcceptedCmp2","AcceptedCmp1","Education","Marital_Status"], inplace=True, axis=1)

z = 0
y = 0
for x in range (0, len(df_mod.columns)):
    sns.boxplot(data=df_mod, x=df_mod.columns[x], ax=axes[z,y])
    y += 1
    if y == 3:
        y = 0
        z += 1
    
plt.show()
```

Now we treat with the more conflicting variables with so many outliers

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 -(1.5 * IQR)
upper_bound = Q3 +(1.5 * IQR)
outliers = (df > upper_bound).sum() + (df < lower_bound).sum()
print(outliers)
```

Reemplazamos los valores de los outliers utilizando la mediana ya que debido a la gran cantidad de datos por fuera de los límites no se nos puede permitir eliminarlos ya que la proporción es muy alta para la cantidad de datos que tenemos disponibles

```python
columns_to_replace = ["MntFishProducts","Income","MntFruits","MntGoldProds","MntMeatProducts","MntSweetProducts","MntWines"
                      ,"NumCatalogPurchases","NumDealsPurchases","NumWebPurchases","NumWebVisitsMonth","Year_Birth"]

for column in columns_to_replace:
    df.loc[df[column] > upper_bound[column], column] = df[column].median()
    df.loc[df[column] < lower_bound[column], column] = df[column].median()
```

```python
outliers = (df > upper_bound).sum() + (df < lower_bound).sum()
print(outliers)
```

There we have the outliers corrected!

# 4. Análisis de datos faltantes 

En esta sección se buscará identificar datos nulos o faltantes en nuestro dataset. Para ello veamos qué features contienen valores nulos

```python
df.info()
```

Dado que contamos con **2240** registros, la única feature que presenta datos faltantes es `Income`

Para definir el método de manejo de datos faltantes, primero se analizará la distribución de `Income`: 

```python
sns.histplot(data=df,
            x='Income',
            kde=True)
```

Dada la distribución de los datos se puede emplear la mediana como un estadístico útil pare rellenar los valores nulos.

```python
median = df['Income'].median()
median
```

```python
df['Income'].fillna(median, inplace=True)
```

Volvemos a consultar si existen valores nulos:

```python
df.info()
```

Como se puede ver, ningún feature cuenta ya con valores nulos.
