## 3. Modelo
### 3.1 Creación del modelo

```python
from sklearn.linear_model import LinearRegression

lin_reg_model = LinearRegression()
```

Para la creacion del modelo generaremos una nueva columna con la cantidad de jovenes y la cantidad de niños por hogar, consolidandola en una nueva variable "YoungHome" 

```python
df['YoungHome'] = df['Kidhome'] + df['Teenhome']
df
```

Del mismo modo, como deseamos saber si la cantidad de jovenes y niños en el tiene una relacion con la cantidad de dinero gastado, consolidamos las cantidades de dinero gastado en cada categoria y se plazman en una nueva columna llamada "MntTotal"

```python
df['MntTotal'] = df['MntFishProducts'] + df['MntMeatProducts'] + df['MntFruits'] + df['MntSweetProducts'] + df['MntWines']
+ df['MntGoldProds']
df
```

Realizamos un diagrama de correlación para poder determinar que variables tiene correlacion entre ellas.

Desde aqui ya podemos identificar que nuestras dos variables creadas no tienen mucha correlacion.

```python
#Diagrama de correlacion completo
plt.figure(figsize=(20,20)) 
ax = sns.heatmap(df.corr(),annot=True,cmap='RdYlGn') 
plt.show()
```

### 3.2 Preaparación y Particionamiento de los datos

```python
from sklearn.model_selection import train_test_split

y = df['MntTotal'].values
X = df['YoungHome'].values

print(y.shape)
print(X.shape)
```

```python
y = y.reshape(-1, 1)
X = X.reshape(-1, 1)
print(y.shape)
print(X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)

plt.scatter(X_train,y_train, label='Raw Data')
plt.title('YoungHome VS MntTotal [TRAIN]')
plt.xlabel('Young in home')
plt.ylabel('MntTotal')
plt.legend()
plt.show()

plt.scatter(X_test,y_test, label='Raw Data')
plt.title('YoungHome VS MntTotal [TEST]')
plt.xlabel('Young in home')
plt.ylabel('MntTotal')
plt.legend()
plt.show()
```

### 3,3 Entrenamos el modelo

```python
lin_reg_model.fit(X_train,y_train)
print('intercept [w0]=', lin_reg_model.intercept_)
print('slope [w1]=', lin_reg_model.coef_)
```

## 4. Evaluación

### 4.1 Evaluar el modelo en el conjunto de entrenamiento

```python
y_pred_train = lin_reg_model.predict(X_train)

plt.scatter(X_train, y_train, label='Real')
plt.plot(X_train,y_pred_train,'r', label='Predicted')
plt.title('YoungHome VS MntTotal')
plt.xlabel('YoungHome')
plt.ylabel('MntTotal')
plt.legend()
plt.show()

r_sq = lin_reg_model.score(X_train, y_train)
print('coefficient R^2 for training data: %.4f' % r_sq)

e_train = y_train - y_pred_train

plt.scatter(y_pred_train,e_train)
a = np.zeros(len(e_train))
a = a.reshape(-1,1)
plt.scatter(y_pred_train,a)
plt.title('Training Data: Residuals vs Predicted values')
plt.ylabel('Residuals [cm]')
plt.xlabel('Predicted [cm]')
plt.show()

mse = metrics.mean_squared_error(y_train, y_pred_train)
rmse = np.sqrt(mse)
print('MSE= %.4f'% mse)
print('RMSE: %.4f' % rmse)
```
