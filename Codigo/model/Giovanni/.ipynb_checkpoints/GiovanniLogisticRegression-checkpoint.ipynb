Importamos librerías importantes que nos ayudarán a la creación del modelo y la creación de las métricas

```python
#Librerías de interés.
import pandas as pd                  #Importamos la librería pandas. Nos va a servir para leer y manipular conjuntos de datos tabulares.
import matplotlib.pyplot as plt      #Importamos pyplot de librería matplotlib. Lo vamos a utilizar para graficar.
import seaborn as sns                #Importamos la librería Seaborn. La vamos a utilizar para graficar.
import numpy as np                   #Importamos la librería numpy para manipular arreglos. 


from sklearn.model_selection import train_test_split #Útil para dividir los conjuntos de datos. 
from sklearn.preprocessing import MinMaxScaler       #Útil para escalar los atributos de entrada.
from sklearn.preprocessing import StandardScaler
from sklearn import metrics    #Permite calcular algunas métricas de desempeño.
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

from copy import deepcopy                            #Permite hacer copias profundas. 
```

Leemos el set de datos e imprimimos la información básica de este, cabe recordar que este dataset tuvo su debido proceso de limpieza y la etapa anterior

```python
df = pd.read_csv("../../datos/clean_marketing_campaing.csv")
df.info()
```

Visualizamos la variable Education para poder cambiar su valor a una variable numérica

```python
df["Education"].value_counts()
```

```python
df["Education"].replace({"Graduation":1, "PhD":2, "Master":3, "2n Cycle":4, "Basic":5}, inplace=True)
df["Education"].value_counts()
```

Visualizamos la variable Marital_Status para poder cambiar su valor a una variable numérica

```python
df["Marital_Status"].value_counts()
```

```python
df["Marital_Status"].replace({"Married":1, "Together":2, "Single":3, "Divorced":4, "Widow":5, "Alone":6, "Absurd":7, "YOLO":8}, inplace=True)
df["Marital_Status"].value_counts()
```

Creamos el mapa de correlación para las variables presentes en el set de datos

```python
fig, ax = plt.subplots(figsize=(15, 15))

sns.heatmap(df.corr(), annot=True, ax=ax)
plt.show()
```

Como vemos el mapa de correlación es demasiado extenso y complicado de interpretar, pero de toda la información que se nos presenta podemos decir que las variables que manejan una mayor correlación con nuestra variable Target son:
* Kidhome, Teenhome, Recency, MntWines, AcceptedCmp3, AcceptedCmp5, AcceptedCmp1, AcceptedCmp2

Sin embargo para el modelo inicial empezaremos probando con la totalidad de variables para luego empezar a probar distintas combinaciones <br> Separamos la variable target de las demás para revisar su distribución

```python
variable_target = "Response"
df_y = df[variable_target]

count_y_values = df_y.value_counts()

count_y_values.plot(kind='pie', figsize=(5, 5),
                                autopct=lambda pct: '{:.1f}%\n({:d})'.format(pct, int(pct/100.*count_y_values.sum())), 
                                startangle=90,       
                                fontsize=14,              
                                cmap='Pastel1')      

plt.title('Target variable distribution along the Dataset') 
plt.ylabel('')
plt.show()
```

Evidenciamos que tenemos una inmensa desproporcionalidad en el set de datos por lo tanto debemos intentar compensar esta diferencia al momento de entrenar los datos, otra técnica interesante para ayudarle al modelo es incrementar la data para el label (1) ya sea recopilando más datos o creando sintéticos a partir de los que ya tenemos <br> Procedemos a eliminar las variables no relevantes en el set de datos (En el caso de la fecha pienso que se podría cambiar por una diferencia de días para obtener una variable más y que sea numérica)

```python
df.drop(["Response","Dt_Customer","ID"], axis=1, inplace=True)
df_x = df
```

Particionamos los datos

```python
seed_value = 5

df_x_train, df_x_test, df_y_train, df_y_test = train_test_split(df_x, df_y, test_size=0.1, 
                                                                random_state=seed_value, 
                                                                shuffle=True)

print("-----------------------------------------------------")
print("Train set size: ", str(df_x_train.shape))
print("-----------------------------------------------------\n")
print("-----------------------------------------------------")
print("Test set size: ", str(df_x_test.shape))
print("-----------------------------------------------------\n")
```

Realizamos un escalado de los datos para la creación del modelo, tener datos normalizados

```python
scale=StandardScaler()
df_x_train = scale.fit_transform(df_x_train)
df_x_test = scale.fit_transform(df_x_test)
```

Iniciamos con la creación del modelo, se eligió una regresión logística ya que tenemos nuestras dos labels definidas claramente, es decir, es un problema de clasificación binaria

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

classifier_model = LogisticRegression(max_iter=500, class_weight={0:1,1:2.4})
```

Procedemos con una pequeña validación cruzada y saber cual puede ser el máximo rendimiento que podremos alcanzar con nuestro modelo propuesto en el fragmento de código anterior

```python
score = cross_val_score(classifier_model, df_x_train, df_y_train, cv=10, scoring='accuracy')
print('Accuracy = {:.4f} and std = {:.4f}'.format(score.mean(),score.std()))
```

Observamos que tenemos un buen desempeño a pesar de lo inmensamente desbalanceado que está el set de datos, esto lo conseguimos incrementando el peso del label (1) ya que como se ve, está establecido que sus calculos pesen 2,5 veces más que los del (0) <br> Ahora procedemos a entrenar el modelo

```python
classifier_model.fit(df_x_train, df_y_train)
```

Una vez con el modelo entrenado, lo que haremos es sacar las predicciones tanto del set de entrenamiento, como del de prueba para así obtener sus métricas de desempeño lo cual nos dará un mejor punto de vista sobre qué tenemos hasta el momento

```python
y_pred_train = classifier_model.predict(df_x_train)
score_train = classifier_model.score(df_x_train, df_y_train)

y_pred_test = classifier_model.predict(df_x_test)
score_test = classifier_model.score(df_x_test, df_y_test)
```

```python
cm = metrics.confusion_matrix(df_y_train, y_pred_train)
print(cm)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score_train)
plt.title(all_sample_title, size = 15);
```

Como vemos para el ejemplo anterior, el modelo tuvo una precisión del 87% en el set de entrenamiento sin embargo observemos las clasificaciones para el label (1) y en las métricas se discutirá si realmente el modelo es bueno

```python
cm = metrics.confusion_matrix(df_y_test, y_pred_test)
print(cm)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score_test)
plt.title(all_sample_title, size = 15);
```

Evidenciamos que para el set de prueba también tuvo un desempeño del 87% pero nuevamente, si observamos el label (1) tendremos que pasar a sus métricas específicas para saber si realmente es un buen modelo

```python
print("Métricas en el set de entrenamiento\n",classification_report(df_y_train, y_pred_train))
```

```python
print("Métricas en el set de entrenamiento\n",classification_report(df_y_test, y_pred_test))
```

Eventualmente como se sospechaba el modelo tiene una inmensa facilidad para predecir las variables con label (0) por la cantidad de datos que obtuvo para entrenar sin embargo para el label (1) Tiene un desempeño bastante deplorable aunque por encima del 50%, al estar hablando de una campaña se podría decir que lo que podemos predecir con facilidad es si un cliente la rechazará con mucha facilidad o si al menos la tendrá en consideración, de cualquier modo esto representa valor para la empresa ya que se podría saber a qué clientes directamente no ofrecerles la campaña y a cuales la podrían tener en consideración para aceptarla o no con una precisión de aproximadamente 61% hablando del F1-Score
